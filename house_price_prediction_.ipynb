reading the dataset

```python
import pandas as pd
cars = pd.read_csv('/content/car data.csv')

cars.head (10)
```

data analysis

```python
cars.info()
cars.isnull().sum()
```

Preprocessing the year coulmn so it become more usful

```python
cars['Car_Age'] = 2024 - cars['Year']
cars
```

transform categorical data into numerical data

```python
from sklearn.preprocessing import LabelEncoder


encoder = LabelEncoder()

cars['Fuel_Type'] = encoder.fit_transform(cars['Fuel_Type'])
cars['Selling_type'] = encoder.fit_transform(cars['Selling_type'])
cars['Transmission'] = encoder.fit_transform(cars['Transmission'])
cars['Car_Name'] = encoder.fit_transform(cars['Car_Name'])
```

libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

Split Data into Training and Testing

```python
X = cars[['Car_Age', 'Present_Price', 'Fuel_Type', 'Selling_type', 'Transmission','Car_Name','Driven_kms','Owner']]
y = cars['Selling_Price']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

the model

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R2 Score: {r2}')

```
